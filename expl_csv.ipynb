```python
# adding all the trips from from april to uber for both the years
uber_april_june_14 = pd.concat([april_14_uber,may_14_uber,june_14_uber])
uber_april_june_14
```

```python
def lat_lon_frequency_section(uber_april_june_14,id_plot):
    geometry = [Point(xy) for xy in zip(uber_april_june_14['Lon'], uber_april_june_14['Lat'])]
# Converting the id_plot into the required coordinates
    id_plot = id_plot.to_crs("EPSG:4326")
    points_df = gpd.GeoDataFrame(uber_april_june_14, geometry=geometry, crs='EPSG:4326')
    clipped_points = gpd.clip(points_df, id_plot.geometry)

# Perform spatial join between clipped points and map polygons
    joined_data = gpd.sjoin(clipped_points, id_plot, how='left', op='within')

# Count the frequency of points within each section
    frequency_by_section_april_june_14 = joined_data.groupby('LocationID').size().reset_index(name='point_frequency')
    return frequency_by_section_april_june_14
frequency_by_section_april_june_14 = lat_lon_frequency_section(uber_april_june_14,id_plot)
```

```python
heat_map_location_ids_april_june_14 = id_plot[id_plot['LocationID'].isin(frequency_by_section_14['LocationID'])]

heat_map_location_ids_april_june_14['frequency'] = frequency_by_section_april_june_14['point_frequency']
frequency_by_section_april_june_14
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
heat_map_location_ids_april_june_14 = heat_map_location_ids_april_june_14.to_crs("EPSG:4326")
heat_map_location_ids_april_june_14.plot(ax=ax, column='frequency', cmap='OrRd', linewidth=0.8, edgecolor='0.05', legend=True)

# Customize the plot
ax.set_title("UBER Pickups Heatmap in New York City APRIL-JUNE'14")
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```
